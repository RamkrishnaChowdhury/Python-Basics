```python
#31.Write a Python program to perform Linear search.

n = int(input("Enter the number of elements in the list: "))
lst = []

for i in range(n):
    element = int(input(f"Enter element {i+1}: "))
    lst.append(element)

target = int(input("Enter the element to search: "))

found = False
for i in range(n):
    if lst[i] == target:
        print(f"Element {target} found at index {i}")
        found = True
        break

if not found:
    print(f"Element {target} not found in the list.")
```

```python
#32. Write a program that accepts sequence of lines as input and prints the lines after making all characters in the sentence capitalized.

lines = []
while True:
    line = input()
    if line:
        lines.append(line.upper())
    else:
        break

for line in lines:
    print(line)
```
